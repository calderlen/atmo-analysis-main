```python
### Model species lines in planet atmosphere with petitRadTrans ###

# PEPSI spectral range: 383 to 912 nm 
# Chemical species of interest: Fe II

# Set-up atmosphere as radiative transfer object with varying pressure layers:

if instrument == 'PEPSI': 
    if arm == 'blue':
        wav_out = np.arange(4200, 5450, 0.01) 
    elif arm == 'red':
        wav_out = np.arange(6200, 7500, 0.01) 
    else:
        wav_out = np.arange(4200, 7500, 0.01)
if instrument == 'MaroonX': wav_out = np.arange(5000, 9200, 0.01)
if instrument == 'NIRSPEC': wav_out = np.arange(22000, 25000, 0.01)
if instrument == 'ESPRESSO': wav_out = np.arange(3800, 7880, 0.01)
    
#np.arange(10000, 25000, 0.01)


#species = ['TiO_48_Exomol_McKemmish TiO_all_iso_Plez', 'K', 'Na_allard', 'H2O_main_iso', 'CO_main_iso']#['Fe','Fe+', 'Mg']
#species = ['H2O_main_iso', 'CH4_main_iso', 'NH3_main_iso'] #'CO_main_iso',
species=['Fe'] #, 'Fe+', 'Si', 'Ni', 'CaH', 'FeH_main_iso', 'TiO_48_Exomol_McKemmish', 'Ti+', 'VO_ExoMol_McKemmish']
species_name = 'Fe'
if 'Plez' in species_name:
    vel_air = True # No, only true for Plez line lists!! No, always False!!! True if Fe+, Ti+; false if from ExoMol
else:
    vel_air = False
    

#the abundance is the atomic mass of the constituent (to get the mass fraction)
#times the log abundance
#Iron: 55.845
#TiO: 47.867+15.999, simple sum estimate
#species_abundance = (55.845) * (10.**(7.48-12))
vmr = 5.4e-5
if 'Ti' in species_name:
    mass = 47.867
if 'TiO' in species_name:
    mass = 47.867+15.999 #TiO
if 'VO' in species_name:
    mass = 50.9415+15.999
if 'Fe' in species_name:
    mass = 55.845
if 'FeH' in species_name:
    mass = 55.845+1.00784
if 'Si' in species_name:
    mass = 28.085
if 'SiO' in species_name:
    mass = 28.085+15.999
if 'OH' in species_name:
    mass = 15.999+1.00784
if 'H2O' in species_name:
    mass = 15.999+2.*1.00784
if 'Cr' in species_name:
    mass = 51.9961
if 'CaH' in species_name:
    mass = 40.078+1.00784
if 'Ni' in species_name:
    mass = 58.6934
if 'MgH' in species_name:
    mass = 24.305 + 1.00784
if 'Al' in species_name:
    mass = 26.982
if 'CH4' in species_name:
    mass = 12.011 + 1.00784 * 4
if 'CO' in species_name:
    mass = 12.011 + 15.999
if 'CO2' in species_name:
    mass = 12.011 + 15.999 * 2
if 'NH3' in species_name:
    mass = 14.007 + 1.00784 * 3
if species_name == 'O':
    mass = 15.999
if 'template' in species_name:
    mass = 1.0
mmw = 2.33
species_abundance = mass/mmw * vmr #1.679e-06
H2_abundance = 1.008 * 1.
He_abundance = 4.00126 * (10.**(10.925-12))
Na_abundance = 22.99 * 10.**(6.24-12)
K_abundance = 39.0983 * 10.**(5.03-12)
CO_abundance = ((12.0096+15.999)/mmw) * 4.6e-4 #0.000459
H2O_abundance = ((15.999+2.*1.00784)/mmw) * 5.4e-4 #0.0003227
VO_abundance = ((50.9415+15.999)/mmw) *  4.3936304946216005e-09
OH_abundance = ((15.999+1.00784)/mmw) *  2.7244708297109155e-07
CO2_abundance = ((12.0096+15.999 * 2)/mmw) * 3.333845083321582e-08
CH4_abundance = ((12.0096 + 1.00784*4.)/mmw) * 4.6e-4
C2H2_abundance = ((12.0096*2. + 1.00784*2.)/mmw) * 9.382196779441383e-12
HCN_abundance = ((12.0096 + 14.00643 + 1.00784)/mmw) * 0.0003418
NH3_abundance = ((14.00643 + 1.00784*3.)/mmw) * 4.1e-5
e_abundance = ((1./1822.8884845)/mmw) * 6.39863899e-9#1.8e-6 #
H_abundance = (1.0/mmw) * 0.0008355 # 1e-1 #
Hm_abundance = (1.0/mmw) * 2.2073098e-12#1e-8#1e-9 #


total_abundance = species_abundance + H2_abundance + He_abundance + e_abundance+ H_abundance + Hm_abundance
abundances = {}
if species_name != 'template': abundances[species_name] = species_abundance/total_abundance
abundances['H2'] = H2_abundance/total_abundance
abundances['He'] = He_abundance/total_abundance
#abundances['Na_allard'] = Na_abundance/total_abundance
#abundances['K'] = K_abundance/total_abundance
abundances['H2O_main_iso'] = H2O_abundance/total_abundance
abundances['CO_main_iso'] = CO_abundance/total_abundance
#abundances['VO_ExoMol_McKemmish'] = VO_abundance/total_abundance
#abundances['OH_main_iso'] = OH_abundance/total_abundance
#abundances['CO2_main_iso'] = CO2_abundance/total_abundance
abundances['CH4_main_iso'] = CH4_abundance/total_abundance
#abundances['C2H2_main_iso'] = C2H2_abundance/total_abundance
#abundances['HCN_main_iso'] = HCN_abundance/total_abundance
abundances['NH3_main_iso'] = NH3_abundance/total_abundance

abundances['H'] = H_abundance/total_abundance
abundances['H-'] = Hm_abundance/total_abundance
abundances['e-'] = e_abundance/total_abundance

#abundances['Fe'] = 5.3e-5/total_abundance
#abundances['Fe+'] = 2.9e-5/total_abundance
#abundances['Si'] = 2.9e-5/total_abundance
#abundances['Ni'] = 1.8e-6/total_abundance
#abundances['CaH'] = 1.8e-6/total_abundance
#abundances['FeH_main_iso'] = 1.9e-7/total_abundance
#abundances['TiO_48_Exomol_McKemmish'] = 1.3e-7/total_abundance
#abundances['Ti+'] = 8.2e-8/total_abundance
#abundances['VO_ExoMol_McKemmish'] = 4.7e-9/total_abundance

```

```python
#Set up parameeter grids
lambda_low = min(wav_out) # PEPSI blue channels spectral range: 480 nm to 544.1 nm 
lambda_high = max(wav_out)
P0 = 1.0 #1 * 10.**(-9) 
kappa_IR = 0.04 
if 'inverted' in temperature_profile:
    gamma = 30# 50 #50
else:
    gamma = 0.5 
    
if planet_name == 'WASP-18b':
    kappa_IR = 10.**(-2.31)
    gamma = 2.*10.**0.86
T_int = 100 # Says Ji
pressures = np.logspace(-10, 2, 130) # default   
if 'two-point' in temperature_profile:
    P1, P2 = 1e-1, 3.2e-4
    T_high = 6000.

if planet_name == 'WASP-84b' or planet_name == 'TOI-3714b' or planet_name == 'NGTS-11b' or planet_name == 'WASP-189b':
    temperature_profile = 'two-point'

if 'two-point' in temperature_profile:
    P1, P2 = 10.**-1.7, 10.**-3.1
    T_high = 4320.
    T_equ = 2200.

#abundance_profile = abundances[species_name] * np.ones_like(pressures)
```

```python
#We have the flux from the planet, now need it relative to the star
#We'll approximate the spectrum of the star as a blackbody

#from astropy.modeling import models
from astropy import units as u



#bb = models.BlackBody(temperature=Teff*u.K)

#flux_st = bb(wav_pl * u.AA)* 4. * np.pi * u.sr #get in same units as the planetary flux
flux_st = nc.b(Teff, nc.c/(wav_pl/1e8)) * u.erg / (u.cm * u.cm) / u.s / u.Hz
pl_bb = nc.b(T_equ, nc.c/(wav_pl/1e8)) * u.erg / (u.cm * u.cm) / u.s / u.Hz
```

```python
if spectrum_type == 'emission':
    flux_pl = flux_pl * u.erg / (u.cm * u.cm) / u.s / u.Hz

    total_flux = flux_pl * (R_pl * u.cm)**2 + flux_st * (R_host * u.cm)**2
    total_flux /= np.mean(total_flux)

    flux_ratio = flux_pl / flux_st * (R_pl / R_host) **2
    flux_ratio += 1.0
    
    pl_bb_ratio = pl_bb / flux_st * (R_pl / R_host) **2
    pl_bb_ratio += 1.0
    
if spectrum_type == 'transmission':
    flux_ratio = 1.0 - (flux_pl**2 - R_pl**2) / R_host**2
```

```python
if normalization_method == 'polynomial': # or normalization_method == 'absolute':
    
    if normalization_method == 'polynomial': flux_here = flux_ratio[:]
    if normalization_method == 'absolute': flux_here = flux_pl[:].value
        
    polynomial_order = 2
    n_segments = 10
    
    npix = len(wav_pl)
    flx_out = np.ones_like(flux_here)
    
    for j in range (n_segments):
    
        fit1=np.polyfit(wav_pl[j*int(npix/n_segments):(j+1)*int(npix/n_segments)], flux_here[j*int(npix/n_segments):(j+1)*int(npix/n_segments)], polynomial_order)

        p = np.polyval(fit1, wav_pl[j*int(npix/n_segments):(j+1)*int(npix/n_segments)])


        for i in range (0,10):
            if i == 0: p2 = p
        
            if spectrum_type == 'emission':
                goods = flux_here[j*int(npix/n_segments):(j+1)*int(npix/n_segments)] <= p2
            if spectrum_type == 'transmission' or planet_name == 'Ji-test':
                goods = flux_here[j*int(npix/n_segments):(j+1)*int(npix/n_segments)] >= p2

            fit2 = np.polyfit(wav_pl[j*int(npix/n_segments):(j+1)*int(npix/n_segments)][goods], flux_here[j*int(npix/n_segments):(j+1)*int(npix/n_segments)][goods], polynomial_order)

            p2 = np.polyval(fit2, wav_pl[j*int(npix/n_segments):(j+1)*int(npix/n_segments)])


        #plt.plot(wav_pl, flux_here/p2)

        flx_out[j*int(npix/n_segments):(j+1)*int(npix/n_segments)] = flux_here[j*int(npix/n_segments):(j+1)*int(npix/n_segments)]/p2
        wav_out = wav_pl
```
